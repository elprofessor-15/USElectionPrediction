```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download NLTK resources
nltk.download('stopwords')
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to clean text
def clean_text(text):
    if pd.isnull(text):  # Handle None or NaN values
        return ""
    text = re.sub(r'@\w+', '', text)  # Remove mentions (e.g., @JoeBiden)
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove punctuation and numbers
    text = text.lower()  # Convert to lowercase
    stop_words = set(stopwords.words('english'))  # Set of stopwords
    text = ' '.join(word for word in text.split() if word not in stop_words)  # Remove stopwords
    return text

# Function to determine sentiment
def get_sentiment(text):
    if text == "":  # If cleaned text is empty
        return "Neutral"
    scores = sia.polarity_scores(text)  # Get sentiment scores
    if scores['compound'] >= 0.05:
        return "Positive"
    elif scores['compound'] <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Function to process the dataframe
def process_dataframe(df):
    if df is None or 'text' not in df.columns:  # Validate DataFrame and column
        raise ValueError("Input DataFrame is invalid or does not contain 'text' column.")
    df['cleaned_text'] = df['text'].apply(clean_text)  # Clean text
    df['sentiment'] = df['cleaned_text'].apply(get_sentiment)  # Get sentiment
    return df

# Process the DataFrame
df1_processed = process_dataframe(df1)
df2_processed = process_dataframe(df2)

# Display the results
print(df1_processed[['user', 'text', 'cleaned_text', 'sentiment']])
print(df2_processed[['user', 'text', 'cleaned_text', 'sentiment']])
```
